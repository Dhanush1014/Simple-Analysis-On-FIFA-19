# Data Cleaning

```python
data[data.columns[data.isna().any()]].isna().sum()
```

```python
data.drop(data[data['Stamina'].isna()].index.tolist(),inplace = True)
```

```python
columns = ['Unnamed: 0','Photo','Flag','Club Logo','Release Clause','Nationality','ID','Club']
try:
    data.drop(data.columns[18:54],axis=1,inplace=True)
    data.drop(columns,axis=1,inplace=True)
except Exception as e:
    print(e)
```

```python
data["Wage"] = data["Wage"].str.replace("€","")
data["Wage"] = data["Wage"].str.replace("M","")
data["Wage"] = data["Wage"].str.replace("K","").astype("float")
```

```python
data["Value"] = data["Value"].str.replace("€","")
data["Value"] = data["Value"].str.replace("M","")
data["Value"] = data["Value"].str.replace("K","").astype("float")
```

# Modelling to predict Overall

```python
data.drop(['Name'],axis=1,inplace=True)
```

```python
sns.heatmap(data.corr().loc[['Overall'],:])
```

```python
data = pd.get_dummies(data)
```

```python
X = data.drop(['Overall'],axis=1)
y = data[['Overall']]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
lm = LinearRegression(normalize=True)
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
score = r2_score(y_test,y_pred)
print(score)
```
